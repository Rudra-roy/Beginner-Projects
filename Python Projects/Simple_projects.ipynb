***Triangle/ Quadrilateral Detector using Angles***

```python

# Asking for numbers of side
sides=int(input("Enter numbers of side: "))
angles=[]
if(sides==3):
    print("Enter the angles:")
    for i in range(0,3):
        n=int(input())
        angles.append(n)
    if(sum(angles)==180):
        if(angles[0]==angles[1]==angles[2]):
            print("A equilateral triangle can be drawn.")
        elif(angles[0]==angles[1] or angles[1]==angles[2] or angles[0]==angles[2]):
            if 90 in angles:
                print('A right angle isoscels triangle can be drawn.')
            elif(angles[0]>90 or angles[1]>90 or angles[2]>90):
                print('A obtuse isoscele triangle can be drawn.')
            else:
                print('A acute isoscele triangle can be drawn.')
        elif(angles[0]!=angles[1] or angles[1]!=angles[2] or angles[0]!=angles[2]):
            if 90 in angles:
                print('A right angle scalene triangle can be drawn.')
            elif(angles[0]>90 or angles[1]>90 or angles[2]>90):
                print('A obtuse scalene triangle can be drawn.')
            else:
                print('A acute scalene triangle can be drawn.')
    else:
        print("Invalid input")
elif(sides==4):
    print("Enter the angles:")
    for i in range(4):
        n=int(input())
        angles.append(n)
    if(sum(angles)==360):
        if angles[0]==angles[1]==angles[2]==angles[3]:
            print("A square or a rectangle can be drawn")
        elif angles[0]==angles[2] and angles[1]==angles[3]:
            print("A rhombus or a parallelogram can be drawn")
    else:
            print("Invalid input")
else:
    print("Invalid input")
```
